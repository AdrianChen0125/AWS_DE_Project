```python
## find the target file order and split based on the years
import os
import pandas as pd

data_folder = "./olist"
raw_data_files = os.listdir(data_folder)  # List all files in the directory

# Iterate through all files in the folder
for filename in raw_data_files:
    parent_folder = "Processed_data"
    
    if "order" in filename.lower():  # Check if filename contains "order"
        
        # Create a directory based on the filename (without prefix and extension)
        
        folder_name = os.path.join(parent_folder, filename[6:-4])
        os.makedirs(folder_name, exist_ok=True)

        # Handle dataset which is not from the main order table
        if "olist_order_" in filename.lower():
            
            # Load the CSV file
            input_file = f'{data_folder}/{filename}'
            df1 = pd.read_csv(input_file)

            # Load the orders dataset
            df2 = pd.read_csv(f'{data_folder}/olist_orders_dataset.csv')[['order_id', 'order_purchase_timestamp']]

            # Merge the dataframes on 'order_id'
            newdf = df1.merge(df2, how='left', on='order_id')

            # Convert 'order_purchase_timestamp' to datetime and extract the year
            newdf['order_purchase_timestamp'] = pd.to_datetime(newdf['order_purchase_timestamp'])
            newdf['year'] = newdf['order_purchase_timestamp'].dt.year

            # Save the split files by year
            for year in newdf['year'].unique():
                output_file = f'./{folder_name}/{filename[6:-4]}_{year}.csv'
                newdf[newdf['year'] == year].to_csv(output_file, index=False)

        else:
            # If it's not an "olist_order_" dataset, just process it normally
            newdf = pd.read_csv(f'{data_folder}/{filename}')

            # Convert 'order_purchase_timestamp' to datetime and extract the year
            newdf['order_purchase_timestamp'] = pd.to_datetime(newdf['order_purchase_timestamp'])
            newdf['year'] = newdf['order_purchase_timestamp'].dt.year

            # Save the split files by year
            for year in newdf['year'].unique():
                output_file = f'./{folder_name}/{filename[6:-4]}_{year}.csv'
                newdf[newdf['year'] == year].to_csv(output_file, index=False)
    else:
        
        output_file = f'./{parent_folder}/{filename[6:-4]}.csv'
        
        newdf = pd.read_csv(f'{data_folder}/{filename}').to_csv(output_file, index=False)
```
